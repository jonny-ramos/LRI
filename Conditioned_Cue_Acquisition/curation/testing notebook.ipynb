```python
def partial_match(s, target):
    str_list = s.split('_')
    for string in str_list:
        if target in string:
            return string
    
partial_match(fs[0], 'rat')
```

```python
arrs = [(f, np.load(f, allow_pickle=True)) for i, f in enumerate(fs) if 'coke' in f and 'PFC' in f]
```

```python
def flatten(kv):
    '''
    kv is a key, value pair tuple where the value is an n by m array
    returns a new list of key value pairs where each value is a 1 by m subarray
    from the original n by m value array. 
    key is unchanged for all new key, value pairs
    '''
    key, value = kv

    return [(key, subarray) for subarray in value]

flattened_arr = []
for arr in arrs:
    f = flatten(arr)
    f = [('_'.join([partial_match(kv[0], 'Ephys'), partial_match(kv[0], 'rat'), f'trial{i}']), kv[1]) for i, kv in enumerate(f)]
    flattened_arr += f
```
