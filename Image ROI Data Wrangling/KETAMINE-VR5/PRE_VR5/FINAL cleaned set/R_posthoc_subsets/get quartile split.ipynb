```python
import numpy as np
import pandas as pd
import os
```

```python
fs = [f for f in os.listdir() if 'split' in f and '.csv' in f]

for f in fs:
    df = pd.read_csv(f)
    
    try:
        df['quartile'] = pd.qcut(df.cFos_intensity, 4, ['q1', 'q2', 'q3', 'q4'])
        df['q4'] = df.quartile == 'q4'
        df = df.rename(columns={'treat':'react', 'react':'treat'})
    except:
        df['quartile'] = pd.qcut(df.Npas4_intensity, 4, ['q1', 'q2', 'q3', 'q4'])
        df['q4'] = df.quartile == 'q4'
        df = df.rename(columns={'treat':'react', 'react':'treat'})
        
    df.to_csv(f)
```
