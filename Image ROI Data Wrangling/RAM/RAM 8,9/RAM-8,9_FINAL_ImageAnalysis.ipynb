```python
df_ram8 = pd.read_csv('RAM-8_raw_data.csv')
print(df_ram8.columns)
print(df_ram8.shape)
df_ram8.head()
```

```python
df = pd.concat([df_ram8, df_ram9])
print(df.columns)
print(df.shape)
df.head()
```

```python
df.columns = [col.lower() for col in df.columns]
df['treatment'] = df[['dox', 'react']].aggregate('_'.join, axis=1)
df['image_name'] = df.filename.apply(lambda x: '_'.join(x.split('_')[:-1]))
df['image_name'] = df[['rat_n', 'image_name']].aggregate('_'.join, axis=1)
df
```

```python
df['rat_n'] = df.rat_n.apply(lambda x: 'RAM-8_'+x if 'C' in x else 'RAM-9_'+x)
df
```

# check for duplicates, nulls

```python
subset = [x.lower() for x in ['XM', 'YM', 'directory', 'roiName', 'FileName', 'Mean']]
df = df.drop_duplicates(subset=subset).copy()
print(df.duplicated(subset=subset).sum())
print(df.shape)
df.head()
```

```python
# build cols we still need
df['mean-background'] = df['mean'] - df['background']
df.head()
```

### standardizing stain labels

```python
df['stain'] = df.stain.apply(lambda x: x.replace('c-Fos', 'cFos')\
    .replace('with', 'w')\
    .replace('/','')\
    .replace('GFP','EGFP')\
    .replace('EEGFP', 'EGFP')\
    .replace('Triple-labeled', 'triple')\
    .replace('single_',''))

print(np.unique(df.stain))
```

```python
print(df.shape)
df = df[df.rat_n != 'RAM-8_rat7C']
print(df.shape)
df.head()
```

```python
cFos_min = abs(df[df.stain.str.contains('cFos')]['mean-background'].min())
mKate_min = abs(df[df.stain.str.contains('mKate2')]['mean-background'].min())
EGFP_min = abs(df[df.stain.str.contains('EGFP')]['mean-background'].min())

print(cFos_min)
print(mKate_min)
print(EGFP_min)

def adjust_intensity(x):
    if 'cFos' in x.stain:
        return x['mean-background'] + cFos_min
    elif 'mKate2' in x.stain:
        return x['mean-background'] + mKate_min
    elif 'EGFP' in x.stain:
        return x['mean-background'] + EGFP_min
    else:
        print('missing val')
        return np.nan

df_adjusted = df.copy(deep=True)
df_adjusted['mean-background'] = df.apply(adjust_intensity, axis=1)
```

# Normalizing Intensity to OFF_DOX FR1

```python
for stain in df_adjusted.stain.unique():
    print(stain)
    df_q = df_adjusted.query(f'stain == "{stain}"').copy()
    norm_condition = 'DOX_OFF_FR1'
    df_norm_condition = df_q.query(f'treatment == "{norm_condition}"')
    norm_mean = df_norm_condition['mean-background'].mean()
    df_q['norm_mean-background'] = df_q['mean-background'] / norm_mean
    print(df_q.groupby('treatment')['norm_mean-background'].mean(),'\n')

    s = stain.replace('/','').replace('-','')

    prism_reorg(df_q).to_csv(f'RAM-8,9_norm_{s}_PRISM.csv')
```

# Counting Mean Cell ns

```python
df_img_ns = df.groupby('rat_n')['image_name'].unique().reset_index()
df_img_ns['n_imgs'] = df_img_ns.image_name.apply(len)
df_img_ns
```

### All stain-type combinations

```python
def mean_cell_ns_single(stain_x):
    # series of groupby's and reductions to get sum of sums per rat
    # first groupby to count cells per image; second groupby to sum counts per rat
    # merge with df to get required labels, merge with df_img_ns to get n_imgs per rat
    q = df_adjusted.query(f'stain == "{stain_x}"')\
        .groupby('image_name')['rat_n'].count()\
        .reset_index().rename(columns={'rat_n':'count'})\
        .merge(df_adjusted[['image_name', 'rat_n']], on='image_name', how='left')\
        .drop_duplicates()\
        .groupby('rat_n')['count'].sum().reset_index().rename(columns={'count':'sum'})\
        .merge(df_adjusted[['rat_n', 'react', 'dox']],\
            on='rat_n', how='left')\
        .drop_duplicates().reset_index()\
        .merge(df_img_ns[['rat_n', 'n_imgs']], on='rat_n', how='left')

    # compute means
    q['mean_cell_n'] = q['sum'] / q['n_imgs']

    # sort by dox then react; reorder cols
    q = q.sort_values(by=['dox', 'react'])\
        [['rat_n', 'dox', 'react', 'sum', 'n_imgs', 'mean_cell_n']]\
        .reset_index().drop('index', axis=1)

    return q

df_total_cFos = mean_cell_ns_single('cFos')

# these should be identical, but i'm performing both queries for completeness
mean_cell_ns_single('cFos').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_cFos_mean_cell_ns.csv')
mean_cell_ns_single('mKate2').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_mKate_mean_cell_ns.csv')
mean_cell_ns_single('EGFP').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_EGFP_mean_cell_ns.csv')

mean_cell_ns_single('cFos_coloc_w_mKate2').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_cFos_coloc_w_mKate2_mean_cell_ns.csv')
mean_cell_ns_single('cFos_coloc_w_EGFP').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_cFos_coloc_w_EGFP_mean_cell_ns.csv')
mean_cell_ns_single('mKate2_coloc_w_cFos').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_mKate2_coloc_w_cFos_mean_cell_ns.csv')
mean_cell_ns_single('mKate2_coloc_w_EGFP').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_mKate2_coloc_w_EGFP_mean_cell_ns.csv')
mean_cell_ns_single('EGFP_coloc_w_cFos').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_EGFP_coloc_w_cFos_mean_cell_ns.csv')
mean_cell_ns_single('EGFP_coloc_w_mKate2').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_EGFP_coloc_w_mKate2_mean_cell_ns.csv')

mean_cell_ns_single('triple_cFos').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_triple_cFos_mean_cell_ns.csv')
mean_cell_ns_single('triple_mKate2').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_triple_mKate2_mean_cell_ns.csv')
mean_cell_ns_single('triple_EGFP').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_triple_EGFP_mean_cell_ns.csv')
```

# splitting hi/lo cFos

```python
cFos_stains = set(['cFos', 'cFos_coloc_w_mKate2', 'cFos_coloc_w_EGFP', 'triple_cFos'])
df_cFos = df_adjusted[df_adjusted.stain.apply(lambda x: x in cFos_stains)].copy(deep=True)
cFos_median = df_cFos['mean-background'].median()
df_cFos['cFos_split'] = df_cFos['mean-background'].apply(lambda x: 'cFos_hi' if x > cFos_median else 'cFos_lo')

print(df_cFos.shape)
print(df_cFos.stain.unique())
df_cFos.head()
```

```python
def mean_cell_ns_cFosSplit(cFos_split, stain_x):
    # series of groupby's and reductions to get sum of sums per rat
    # first groupby to count cells per image; second groupby to sum counts per rat
    # merge with df to get required labels, merge with df_img_ns to get n_imgs per rat
    q = df_cFos.query(f'cFos_split == "{cFos_split}" and stain == "{stain_x}"')\
        .groupby('image_name')['rat_n'].count()\
        .reset_index().rename(columns={'rat_n':'count'})\
        .merge(df_cFos[['image_name', 'rat_n']], on='image_name', how='left')\
        .drop_duplicates()\
        .groupby('rat_n')['count'].sum().reset_index().rename(columns={'count':'sum'})\
        .merge(df_cFos[['rat_n', 'react', 'dox']],\
            on='rat_n', how='left')\
        .drop_duplicates().reset_index()\
        .merge(df_img_ns[['rat_n', 'n_imgs']], on='rat_n', how='left')

    # compute means
    q['mean_cell_n'] = q['sum'] / q['n_imgs']

    # sort by dox then react; reorder cols
    q = q.sort_values(by=['dox', 'react'])\
        [['rat_n', 'dox', 'react', 'sum', 'n_imgs', 'mean_cell_n']]\
        .reset_index().drop('index', axis=1)

    return q

mean_cell_ns_cFosSplit('cFos_hi', 'cFos').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_cFos_hi_mean_cell_ns.csv')
mean_cell_ns_cFosSplit('cFos_lo', 'cFos').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_cFos_lo_mean_cell_ns.csv')

mean_cell_ns_cFosSplit('cFos_hi', 'cFos_coloc_w_mKate2').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_cFos_hi_coloc_w_mKate2_mean_cell_ns.csv')
mean_cell_ns_cFosSplit('cFos_lo', 'cFos_coloc_w_mKate2').merge(df_total_cFos[['rat_n', 'dox', 'react']], on=['rat_n','dox', 'react'], how='right').to_csv('RAM-8,9_cFos_lo_coloc_w_mKate2_mean_cell_ns.csv')
```
