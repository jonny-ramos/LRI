```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

files = [f for f in os.listdir() if '.npy' in f]
files.sort()
files
```

```python
data = [np.load(f, allow_pickle=True) for f in files]
keys = ['VR5_ChABC_HPC_rw', 'VR5_ChABC_HPC_unrw', 'VR5_ChABC_PFC_rw', 'VR5_ChABC_PFC_unrw',
       'VR5_Vehicle_HPC_rw', 'VR5_Vehicle_HPC_unrw', 'VR5_Vehicle_PFC_rw', 'VR5_Vehicle_PFC_unrw']

dct_data = dict(zip(keys, data))
floor, ceil = 1.96, 4
```

```python
ITPCz = dct_data['VR5_ChABC_HPC_rw']

fs = 2000
time = np.arange(-300,1000,1/fs*1000)
freqs = np.arange(4,40)

plt.figure(figsize=(6.5,5))
cs = plt.contourf(time, freqs, ITPCz, levels = np.linspace(floor, ceil, 40), extend='min', cmap='jet')
cbar = plt.colorbar(cs,shrink=0.9)
plt.xlabel('Time (ms), press @ t=0', fontsize=22)
plt.ylabel('Frequency (Hz)', fontsize=22)
plt.title('VR5 ChABC HPC_Rside Rewarded \nZ-scored ITPC, p>=0.05', fontsize=20)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
cbar.ax.tick_params(labelsize=16)
plt.tight_layout()

plt.savefig('VR5 ChABC HPC_Rside Rewarded Z-scored ITPC.png', dpi=600)
```
